```python
import sys
sys.path.append('..')
import pandas as pd
import torch
from dataset.cinic_dataset import CinicDataset
from augmentation.augmentor import Augmentor
from models.cnn import ConvolutionalNeuralNetwork
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
```

```python
df = pd.read_csv("../../data/df_train.csv")
df_test = pd.read_csv("../../data/df_test.csv")
```

```python
# Load the datasets
df_train = df.sample(frac=0.8, random_state=420)
df_val = df.drop(df_train.index)
train_set = CinicDataset(df_train)
val_set = CinicDataset(df_val)
test_set = CinicDataset(df_test)
```

```python
train_loader = DataLoader(train_set, shuffle=True, batch_size=64)
val_loader = DataLoader(val_set, shuffle=True, batch_size=64)
```

```python
model = ConvolutionalNeuralNetwork()
```

```python
n_epochs = 1
augmentor = Augmentor()
for epoch in range(1, n_epochs+1):
    torch.manual_seed(epoch)

    print(f"EPOCH: {epoch}/{n_epochs}")

    train_loss = 0.0
    
    for (batch_idx, sample) in enumerate(train_loader):
        inputs_list = augmentor.augment_data(sample['image'])
        labels = sample['label']
        for input in inputs_list:
            model.fit(input, labels)
            loss = model.loss
            train_loss += loss.item()
        if batch_idx % 10 == 0:
            print(f"Batch: {batch_idx}, Loss: {loss.item()}")
    print(f"Train Loss: {train_loss}")
```
